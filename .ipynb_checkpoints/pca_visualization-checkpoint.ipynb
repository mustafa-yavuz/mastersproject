```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

with open('abstract_10000.pkl', 'rb') as f:
    abstract_10000 = pickle.load(f)

with open('title_list.pkl', 'rb') as f:
    title_list = pickle.load(f)
```

```python
abstract_10000_np = list(map(lambda x: x.cpu().detach().numpy(), abstract_10000))
```

```python
# Creating dataframe with a specific number of clusters

def cluster_df(n_clusters): 

    # PCA with 2 components
    X = np.array(abstract_10000_np)
    pca = PCA(n_components=2)
    X_reduced = pca.fit_transform(X)
    label_list = []

    # Clustering
    km = KMeans(n_clusters=n_clusters)
    km.fit(X_reduced)
    label_list.append(km.labels_)

    # Creating dataframe
    pca_df = pd.DataFrame(index=title_list,columns=['First Component','Second Component'],data=X_reduced)
    pca_df[f'k={n_clusters}'] = np.array(label_list).T

    #pca_df['IDs'] = np.arange(1,10000)

    return pca_df
```

```python
# Trying the method with 3 clusters and checking head of the dataframe
pca_df = cluster_df(3)
pca_df.head()
```

```python
# Plotting the samples for 3 clusters
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.scatterplot(x=pca_df['First Component'],y=pca_df['Second Component'],hue=pca_df.iloc[:,2])
```

```python
pca_df['k=5'] = cluster_df(5).iloc[:,2]
pca_df['k=10'] = cluster_df(10).iloc[:,2]
pca_df.head()
```

```python
# Plotting the samples for 5 clusters
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.scatterplot(x=pca_df['First Component'],y=pca_df['Second Component'],hue=pca_df.iloc[:,3],palette ='Dark2')
```

```python
# Plotting the samples for 10 clusters
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.scatterplot(x=pca_df['First Component'],y=pca_df['Second Component'],hue=pca_df.iloc[:,4],palette ='Dark2')
```
